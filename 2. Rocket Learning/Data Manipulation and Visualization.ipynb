### Custom Binning

```python
# 40 Days
# Very Active      -    >3 per Day (40*3=120) 
# Active           -    >1 per Day (120-40)
# Okay/Neutral          -    <1 per Day (40-30)      
# Inactive         -    once per (2 Day- 4 Day) (30-5) (missing month or 2)     
# Very Inactive    -    (5-0) (attending only one event/ most probably missed the summer camp)

#Subset- Percentage who missed the summer camp vs who attended
```

```python

def custom_bins(counts):
    if counts>=120:
        label='Very Active'
    elif counts<120 and counts>40:
        label='Active'
    elif counts<40 and counts>30:
        label='Okay'
    elif counts<30 and counts>5:
        label='Inactive'
    else:
        label='Very Inactive'
        
    return label   
```

```python
df['bins_cust']=df['counts'].apply(custom_bins)
```

```python
df
```

```python
df['bins_cust'].unique()
```

```python
mapp=dict(zip(df.index,df['bins_cust']))
mapp
```

```python
data1['parent_activity_cust']=data1['parent_ID'].map(mapp)
```
